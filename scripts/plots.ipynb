```python
import json
import numpy as np
import matplotlib.pyplot as plt


def extract_metrics(json_file):
    with open(json_file) as f:
        data = json.load(f)
    labels = data['details']
    
    t1w_dice = []
    t1w_assd = []
    t2w_dice = []
    t2w_assd = []
    
    for label in labels.values():
        for file_info in label['files']:
            filename = file_info['filename']
            dice = file_info['dice']
            assd = file_info['assd']
            if 'T1w' in filename:
                t1w_dice.append(dice)
                t1w_assd.append(assd)
            elif 'T2w' in filename:
                t2w_dice.append(dice)
                t2w_assd.append(assd)
    
    return (t1w_dice, t1w_assd, t2w_dice, t2w_assd)


json_files = [
    '/Users/yunanli/Desktop/Infresults/Inf35_modelres_250/Inf35_modelres_250.json',
    '/Users/yunanli/Desktop/Infresults/Inf34_modelres_250/Inf34_modelres_250.json',
    '/Users/yunanli/Desktop/Infresults/Inf36_modelres_250/Inf36_modelres_250.json',
    '/Users/yunanli/Desktop/Infresults/Inf37_modelres_250/Inf37_modelres_250.json',
    '/Users/yunanli/Desktop/Infresults/Inf38_modelres_250/Inf38_modelres_250.json'
]



all_t1w_dice = []
all_t1w_assd = []
all_t2w_dice = []
all_t2w_assd = []


for json_file in json_files:
    t1w_dice, t1w_assd, t2w_dice, t2w_assd = extract_metrics(json_file)
    all_t1w_dice.append(t1w_dice)
    all_t1w_assd.append(t1w_assd)
    all_t2w_dice.append(t2w_dice)
    all_t2w_assd.append(t2w_assd)

t1w_dice_avg = [np.mean(dice) for dice in all_t1w_dice]
t1w_dice_std = [np.std(dice) for dice in all_t1w_dice]
t1w_assd_avg = [np.mean(assd) for assd in all_t1w_assd]
t1w_assd_std = [np.std(assd) for assd in all_t1w_assd]

t2w_dice_avg = [np.mean(dice) for dice in all_t2w_dice]
t2w_dice_std = [np.std(dice) for dice in all_t2w_dice]
t2w_assd_avg = [np.mean(assd) for assd in all_t2w_assd]
t2w_assd_std = [np.std(assd) for assd in all_t2w_assd]


x_labels = ['2x2x2', '1x1x1', '0.7x0.7x0.7', '0.6x0.6x0.6', '0.5x0.5x0.5']
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.errorbar(x_labels, t1w_dice_avg, yerr=t1w_dice_std, label='T1w Dice Score', fmt='-o', color='b')
ax1.errorbar(x_labels, t2w_dice_avg, yerr=t2w_dice_std, label='T2w Dice Score', fmt='-x', color='g')
ax1.set_ylabel('Dice Score', color='b')
ax1.set_ylim(0.8, 1)
ax1.set_xticks(np.arange(len(x_labels)))
ax1.set_xticklabels(x_labels, rotation=45, ha='right')

ax2 = ax1.twinx()
ax2.errorbar(x_labels, t1w_assd_avg, yerr=t1w_assd_std, label='T1w ASSD', fmt='-o', color='r')
ax2.errorbar(x_labels, t2w_assd_avg, yerr=t2w_assd_std, label='T2w ASSD', fmt='-x', color='m')
ax2.set_ylabel('ASSD', color='r')
plt.title('Comparison of Dice Score and ASSD for T1w and T2w on model resolution')
fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
plt.show()
```

## Evaluation for the difference of DICE and ASSD for the difference between T1w and T2w

```python
import json
import matplotlib.pyplot as plt

class plot_t1_t1_labeldiff:

    def extract_metrics(self,json_file):
        with open(json_file, 'r') as file:
            data = json.load(file)

        t1w_data = {}
        t2w_data = {}

        for label in data['details']:
            label_data = data['details'][label]['files']
            for item in label_data:
                filename = item['filename']
                subject_id = filename.split('_')[0]
                modality = filename.split('_')[1].split('.')[0]
                
                if modality == 'T1w':
                    if subject_id not in t1w_data:
                        t1w_data[subject_id] = {}
                    t1w_data[subject_id][label] = {'dice': item['dice'], 'assd': item['assd']}
                elif modality == 'T2w':
                    if subject_id not in t2w_data:
                        t2w_data[subject_id] = {}
                    t2w_data[subject_id][label] = {'dice': item['dice'], 'assd': item['assd']}
        
        return t1w_data, t2w_data

    def plot_diff(self,json_file, save_path):
        t1w_data, t2w_data = extract_metrics(json_file)
        dice_differences = {'1': [], '2': [], '3': []}
        assd_differences = {'1': [], '2': [], '3': []}

        for subject_id in t1w_data:
            if subject_id in t2w_data:
                for label in t1w_data[subject_id]:
                    if label in t2w_data[subject_id]:
                        dice_diff = t2w_data[subject_id][label]['dice'] - t1w_data[subject_id][label]['dice']
                        assd_diff = t2w_data[subject_id][label]['assd'] - t1w_data[subject_id][label]['assd']
                        dice_differences[label].append(dice_diff)
                        assd_differences[label].append(assd_diff)

        labels = ['1', '2', '3']
        x = range(len(dice_differences['1']))

        fig, ax1 = plt.subplots(figsize=(14, 7))
        colors = {'1': 'blue', '2': 'green', '3': 'red'}
        ax1.set_xlabel('Subjects')
        ax1.set_ylabel('Dice Difference', color='tab:blue')
        for label in labels:
            ax1.plot(x, dice_differences[label], label=f'Label {label} Dice Diff', marker='o', color=colors[label], alpha=0.6)
        ax1.tick_params(axis='y', labelcolor='tab:blue')
        ax1.set_ylim([-0.2, 0.2])

        ax2 = ax1.twinx()
        ax2.set_ylabel('ASSD Difference', color='tab:red')
        for label in labels:
            ax2.plot(x, assd_differences[label], label=f'Label {label} ASSD Diff', marker='x', color=colors[label], alpha=0.6)
        ax2.tick_params(axis='y', labelcolor='tab:red')
        ax2.set_ylim([-3, 3])

        fig.tight_layout()
        fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
        plt.title('Differences in Dice and ASSD Scores between T1w and T2w Images')
        plt.savefig(save_path)
```

```python
diff_plot = plot_t1_t1_labeldiff()
diff_plot.plot_diff('/Users/yunanli/Desktop/Infresults/Inf38_nativeres_250/Inf38_nativeres_250.json', '/Users/yunanli/Desktop/Infresults/Inf38_nativeres_250/Inf38_nativeres_250_diff.png')
```

## Average ADDS DICE with different resolution within a dataset, I added a function to remove the outliers, this will be only applied to the native resolution dataset, since model resolutions have the same resolution within the dataset.

```python
import json
import numpy as np
import matplotlib.pyplot as plt

def remove_outliers(data):
    data = np.array(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return data[(data >= lower_bound) & (data <= upper_bound)]

def plot_metrics_by_group(json_file_path, groups_dict):
    with open(json_file_path, 'r') as f:
        data = json.load(f)
    
    group_avg_dice = {}
    group_avg_assd = {}
    
    for group, files in groups_dict.items():
        dice_values = []
        assd_values = []
    
        for file in files:
            for detail in data['details'].values():
                for file_detail in detail['files']:
                    if file_detail['filename'].startswith(file):
                        dice_values.append(file_detail['dice'])
                        assd_values.append(file_detail['assd'])
        
        dice_values = remove_outliers(dice_values)
        assd_values = remove_outliers(assd_values)

        if dice_values.size > 0 and assd_values.size > 0:
            group_avg_dice[group] = np.mean(dice_values)
            group_avg_assd[group] = np.mean(assd_values)
        else:
            group_avg_dice[group] = None
            group_avg_assd[group] = None

    groups = list(groups_dict.keys())
    avg_dice = [group_avg_dice[group] for group in groups if group_avg_dice[group] is not None]
    avg_assd = [group_avg_assd[group] for group in groups if group_avg_assd[group] is not None]
    valid_groups = [group for group in groups if group_avg_dice[group] is not None and group_avg_assd[group] is not None]
    
    fig, ax1 = plt.subplots()
    
    ax1.set_xlabel('Groups')
    ax1.set_ylabel('Average Dice', color='tab:blue')
    ax1.plot(valid_groups, avg_dice, color='tab:blue', marker='o', label='Average Dice')
    ax1.tick_params(axis='y', labelcolor='tab:blue')
    
    ax2 = ax1.twinx()
    ax2.set_ylabel('Average ASSD', color='tab:red')
    ax2.plot(valid_groups, avg_assd, color='tab:red', marker='o', label='Average ASSD')
    ax2.tick_params(axis='y', labelcolor='tab:red')
    
    fig.tight_layout()
    plt.title('Average Dice and ASSD by Group (Outliers Removed)')
    plt.show()

json_file_path = '/Users/yunanli/Desktop/Infresults/Inf38_nativeres_250/Inf38_nativeres_250.json'
groups_dict = {
    1: ['sub-0234_acq-iso_T2w.nii', 'sub-0241_acq-iso_T2w.nii', 'sub-0239_acq-iso_T2w.nii', 'sub-0227_T1w.nii', 'sub-0227_T2w.nii', 'sub-0256_T1w.nii', 'sub-0250_T1w.nii'],
    2: ['sub-0253_T1w.nii', 'sub-0253_T2w.nii', 'sub-0250_T2w.nii', 'sub-0256_T2w.nii', 'sub-0245_T1w.nii', 'sub-0245_T2w.nii'],
    3: ['sub-0231_T1w.nii', 'sub-0231_T2w.nii', 'sub-0237_T1w.nii', 'sub-0255_T2w.nii', 'sub-0255_T1w.nii', 'sub-0252_T2w.nii'],
    4: ['sub-0252_T1w.nii', 'sub-0239_T2w.nii', 'sub-0239_T1w.nii', 'sub-0234_T1w.nii', 'sub-0234_T2w.nii', 'sub-0241_T2w.nii'],
    5: ['sub-0241_T1w.nii', 'sub-0244_T2w.nii', 'sub-0244_T1w.nii', 'sub-0243_T2w.nii', 'sub-0243_T1w.nii', 'sub-0251_T1w.nii'],
    6: ['sub-0251_T2w.nii', 'sub-0232_T2w.nii', 'sub-0232_T1w.nii', 'sub-0223_T2w.nii', 'sub-0223_T1w.nii', 'sub-0254_T1w.nii'],
    7: ['sub-0254_T2w.nii', 'sub-0236_T1w.nii', 'sub-0236_T2w.nii', 'sub-0228_T2w.nii', 'sub-0228_T1w.nii', 'sub-0242_T1w.nii'],
    8: ['sub-0242_T2w.nii', 'sub-0229_T1w.nii', 'sub-0229_T2w.nii', 'sub-0257_T2w.nii', 'sub-0257_T1w.nii', 'sub-0237_T2w.nii'],
    9: ['sub-0224_T2w.nii', 'sub-0224_T1w.nii', 'sub-0222_T1w.nii', 'sub-0249_T1w.nii', 'sub-0222_T2w.nii', 'sub-0249_T2w.nii'],
    10: ['sub-0226_T2w.nii', 'sub-0233_T2w.nii', 'sub-0225_T2w.nii', 'sub-0233_T1w.nii', 'sub-0225_T1w.nii', 'sub-0226_T1w.nii']
}

plot_metrics_by_group(json_file_path, groups_dict)
```
